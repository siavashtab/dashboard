```python
import ipywidgets as widgets
from IPython.display import display

def plot_stock_data(ticker_symbol):
    print(f"Fetching data for {ticker_symbol}...")
    stock_yf = yf.Ticker(ticker_symbol)
    data = stock_yf.history(period="1y") # Fetching 1 year of data for sufficient weekly data points

    if data.empty:
        print(f"No data found for {ticker_symbol}")
        return

    # Resample daily data to weekly data
    weekly_data = data.resample('W').agg({
        'Open': 'first',
        'High': 'max',
        'Low': 'min',
        'Close': 'last',
        'Volume': 'sum'
    })

    # Calculate 10-week and 20-week moving averages on the 'Close' price
    weekly_data['MA10'] = weekly_data['Close'].rolling(window=10).mean()
    weekly_data['MA20'] = weekly_data['Close'].rolling(window=20).mean()

    # Create a list of 'addplot' objects for the moving averages
    apds = [
        mpf.make_addplot(weekly_data['MA10'], color='blue', panel=0, width=0.7, linestyle='dashed'),
        mpf.make_addplot(weekly_data['MA20'], color='red', panel=0, width=0.7, linestyle='dashed')
    ]

    stock_name = sp500_ticker_symbols.get(ticker_symbol, ticker_symbol) # Get name, or use symbol if not found

    # Plot the candlestick chart with moving averages
    fig, axes = mpf.plot(weekly_data,
                         type='candle',
                         style='yahoo',
                         title=f"{stock_name} ({ticker_symbol}) Weekly Candlestick with 10 & 20-week MAs",
                         ylabel='Price',
                         ylabel_lower='Volume',
                         volume=True,
                         addplot=apds,
                         returnfig=True)

    mpf.show()
```

```python
# Create a dropdown widget for S&P 500 tickers
sp500_ticker_list = sorted(list(sp500_ticker_symbols.keys()))
ticker_selector = widgets.Dropdown(
    options=sp500_ticker_list,
    value=sp500_ticker_list[0], # Default value
    description='Select Ticker:',
    disabled=False,
)

# Display the interactive dashboard
output = widgets.interactive_output(plot_stock_data, {'ticker_symbol': ticker_selector})
display(ticker_selector, output)
```

```python
import yfinance as yf
import pandas as pd

# Get all S&P 500 ticker symbols from the previously defined dictionary
tickers = list(sp500_ticker_symbols.keys())

# Fetch historical data for all tickers
# Using '5y' for a reasonable period to calculate covariance
all_stock_data = {}
errors = []

for ticker_symbol in tickers:
    try:
        stock = yf.Ticker(ticker_symbol)
        hist_data = stock.history(period="5y")
        if not hist_data.empty:
            all_stock_data[ticker_symbol] = hist_data['Close']
        else:
            errors.append(f"No data found for {ticker_symbol}")
    except Exception as e:
        errors.append(f"Error fetching data for {ticker_symbol}: {e}")

# Combine all 'Close' prices into a single DataFrame
price_data = pd.DataFrame(all_stock_data)

# Drop any columns (tickers) that might be all NaN after alignment
price_data.dropna(axis=1, inplace=True)

print(f"Successfully fetched data for {len(price_data.columns)} out of {len(tickers)} S&P 500 tickers.")
if errors:
    print("Errors encountered for the following tickers:")
    for error in errors:
        print(f"- {error}")

print("First 5 rows of price data:")
display(price_data.head())
```

```python
# Calculate daily returns
returns = price_data.pct_change().dropna()

# Calculate mean daily returns and the covariance matrix
mean_returns = returns.mean()
cov_matrix = returns.cov()

print("Mean Daily Returns (first 5):")
display(mean_returns.head())

print("Covariance Matrix (first 5x5):")
display(cov_matrix.iloc[:5, :5])
```

```python
import cvxpy as cp
import numpy as np

num_assets = len(mean_returns)

# Define optimization variables: portfolio weights
# cp.Variable(num_assets) creates a vector of `num_assets` variables
weights = cp.Variable(num_assets)

# Define portfolio return and risk (variance)
portfolio_return = weights @ mean_returns.values # @ is matrix multiplication
portfolio_risk = cp.quad_form(weights, cov_matrix.values)

# Define the optimization problem

# 1. Objective: Maximize portfolio return for a given risk aversion (lambda)
# We'll use a placeholder for lambda for now. A common approach is to solve for a range of lambdas
# to trace out the efficient frontier.
# Let's set a target risk instead for a simpler example, or a simple risk aversion coefficient.
risk_aversion = 1.0 # Adjust this value to control the trade-off between risk and return
objective = cp.Maximize(portfolio_return - risk_aversion * portfolio_risk)

# 2. Constraints:
#   - Sum of weights must be 1 (fully invested portfolio)
#   - Weights must be non-negative (no short selling)
constraints = [
    cp.sum(weights) == 1,
    weights >= 0
]

# Create the problem and solve it
problem = cp.Problem(objective, constraints)
problem.solve()

# Print the results
print("Optimal Weights:")
display(pd.Series(weights.value, index=mean_returns.index).sort_values(ascending=False).head(10))

print(f"\nMaximized Portfolio Return: {portfolio_return.value:.4f}")
print(f"Minimized Portfolio Risk (Variance): {portfolio_risk.value:.4f}")
```

This code sets up a basic mean-variance optimization problem:

1.  **Variables**: We define `weights` as a `cvxpy` variable to represent the allocation to each asset.
2.  **Portfolio Return**: Calculated as the dot product of the weights and the mean returns of each asset.
3.  **Portfolio Risk (Variance)**: Calculated using the quadratic form of the weights and the covariance matrix.
4.  **Objective**: We aim to maximize a utility function that balances return and risk. Here, `portfolio_return - risk_aversion * portfolio_risk` maximizes return while penalizing risk, with `risk_aversion` controlling how much risk we are willing to take.
5.  **Constraints**: We ensure that the sum of weights equals 1 (the entire portfolio is invested) and that individual weights are non-negative (no short-selling).

The `problem.solve()` function then finds the optimal weights that satisfy these conditions. You can adjust the `risk_aversion` parameter to explore different risk-return trade-offs.

```python
import cvxpy as cp
import numpy as np
import pandas as pd

num_assets = len(mean_returns)

# Ensure covariance matrix is symmetric and positive semi-definite for Cholesky decomposition
# Add a small diagonal perturbation for numerical stability if needed
cov_matrix_np = cov_matrix.values
if not np.all(np.linalg.eigvals(cov_matrix_np) >= 0):
    print("Warning: Covariance matrix is not positive semi-definite. Adding a small regularization term.")
    cov_matrix_np += np.eye(num_assets) * 1e-6

# Perform Cholesky decomposition of the covariance matrix (Sigma = F @ F.T)
# This is used to convert quadratic form w.T @ Sigma @ w to ||F.T @ w||_2^2
F = np.linalg.cholesky(cov_matrix_np)

# Define optimization variables:
# 1. weights: portfolio allocation for each asset
# 2. t: auxiliary variable representing the portfolio standard deviation (risk)
weights = cp.Variable(num_assets)
t = cp.Variable()

# Define objective: Minimize portfolio standard deviation (t)
objective = cp.Minimize(t)

# Define constraints:
# 1. Sum of weights must be 1 (fully invested portfolio)
# 2. Weights must be non-negative (no short selling)
# 3. Portfolio return must be greater than or equal to a target return
# 4. Second-order cone constraint: ||F.T @ weights||_2 <= t

# Let's set a target return. For demonstration, we can pick a value within the range of possible returns.
# For a more complete efficient frontier, you would iterate over a range of target_returns.
# Calculate the maximum possible return to ensure our target is achievable.
max_possible_return = np.max(mean_returns.values)
min_possible_return = np.min(mean_returns.values)
# Set target return as a fraction of the maximum return or an average
target_return = np.mean(mean_returns.values) # Example: average daily return

print(f"Target daily return set to: {target_return:.6f}")

constraints = [
    cp.sum(weights) == 1,
    weights >= 0,
    mean_returns.values @ weights >= target_return,
    cp.norm2(F.T @ weights) <= t # Second-order cone constraint
]

# Create the problem and solve it
problem = cp.Problem(objective, constraints)
problem.solve()

# Print the results
if problem.status == cp.OPTIMAL or problem.status == cp.OPTIMAL_INACCURATE:
    print("\nOptimal Weights (Top 10):")
    display(pd.Series(weights.value, index=mean_returns.index).sort_values(ascending=False).head(10))

    print(f"\nOptimal Portfolio Standard Deviation: {t.value:.4f}")
    print(f"Achieved Portfolio Return: {(mean_returns.values @ weights.value):.4f}")
else:
    print(f"Problem did not solve to optimality. Status: {problem.status}")
```

This formulation explicitly uses a conic program:

1.  **Cholesky Decomposition**: We first perform a Cholesky decomposition (`F = np.linalg.cholesky(cov_matrix_np)`) of the covariance matrix. This allows us to rewrite the portfolio variance `w^T * Sigma * w` as `||F^T * w||_2^2`.
2.  **Variables**: We define `weights` as before, and introduce an auxiliary variable `t` which represents the portfolio's standard deviation.
3.  **Objective**: We aim to `minimize(t)`, which is equivalent to minimizing the portfolio's standard deviation (risk).
4.  **Second-Order Cone Constraint**: The key to the conic formulation is the constraint `cp.norm2(F.T @ weights) <= t`. This is a Second-Order Cone (SOC) constraint, which defines the feasible region as a cone. This constraint mathematically links the portfolio weights to the risk variable `t`.
5.  **Other Constraints**: We maintain the linear constraints for the sum of weights, non-negativity of weights, and crucially, add a constraint that the `mean_returns.values @ weights >= target_return` to ensure the portfolio achieves at least a specified level of return.

This structure allows solvers designed for conic programming (like those used by `cvxpy`) to efficiently find the optimal portfolio allocation that minimizes risk for a given target return.
